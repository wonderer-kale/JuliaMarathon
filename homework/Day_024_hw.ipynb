```julia
using StatsBase, RDatasets
using Gadfly
```

```julia
iris = dataset("datasets", "iris")
```

```julia
p = p = plot(iris, x = "PetalLength", xintercept = [mean(iris[!, :PetalLength])],
    Geom.vline(color = ["black"]), Geom.histogram)
```
